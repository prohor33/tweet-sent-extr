### Looking for intersected answers, but with zero jaccard

```python
count = 0
for idx in range(len(with_errors_df)):
    row = with_errors_df.iloc[idx]
    target = row['target_string_strip']
    output = row['filtered_output_strip']
    tweet = row['original_tweet']
    jac = row['jac']
    if (target in output or output in target) and jac == 0.0:
        print(f'idx: {idx}, jac: {jac}, t vs o \"{target}\" vs \"{output}\", tweet: \"{tweet}\"')
        count += 1
print(f'count: {count}')
```
