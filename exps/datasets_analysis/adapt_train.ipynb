```python
df = pd.DataFrame(df[df['text'].notnull()])
```

```python
def adapt_text(text):
    text = text.replace('&', '&amp;')
    text = text.replace('\'', '&quot;')
    text = text.replace('<', '&lt;')
    text = text.replace('>', '&gt;')
    return text.replace('`', '\'')
```

```python
df['selected_text'] = df['selected_text'].apply(adapt_text)
df['text'] = df['text'].apply(adapt_text)
```
